```python
import pandas as pd 
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from scipy.sparse import csr_matrix
import sparse_dot_topn.sparse_dot_topn as ct  # Leading Juice for us
import time
```

```python
def ngrams(string, n=3):
    string = re.sub(r'[,-./]|\sBD',r'', string)
    ngrams = zip(*[string[i:] for i in range(n)])
    return [''.join(ngram) for ngram in ngrams]
```

```python
# Testing ngrams work for verification
print(target_left['Player'].iloc[0])
ngrams(target_left['Player'].iloc[0])
```

```python
player_names_df = df_left['long_name']
player_names_df = player_names_df.append(target_left['Player'])
player_names_df =player_names_df.reset_index()[0]

vectorizer = TfidfVectorizer(min_df=1, analyzer=ngrams)
tf_idf_matrix = vectorizer.fit_transform(player_names_df)
```

```python
def awesome_cossim_top(A, B, ntop, lower_bound=0):
    # force A and B as a CSR matrix.
    # If they have already been CSR, there is no overhead
    A = A.tocsr()
    B = B.tocsr()
    M, _ = A.shape
    _, N = B.shape
 
    idx_dtype = np.int32
 
    nnz_max = M*ntop
 
    indptr = np.zeros(M+1, dtype=idx_dtype)
    indices = np.zeros(nnz_max, dtype=idx_dtype)
    data = np.zeros(nnz_max, dtype=A.dtype)
    ct.sparse_dot_topn(
        M, N, np.asarray(A.indptr, dtype=idx_dtype),
        np.asarray(A.indices, dtype=idx_dtype),
        A.data,
        np.asarray(B.indptr, dtype=idx_dtype),
        np.asarray(B.indices, dtype=idx_dtype),
        B.data,
        ntop,
        lower_bound,
        indptr, indices, data)
    return csr_matrix((data,indices,indptr),shape=(M,N))
```

```python
#  Top 10 with similarity above 0.8
t1 = time.time()
matches = awesome_cossim_top(tf_idf_matrix, tf_idf_matrix.transpose(), 10, 0.6)
t = time.time()-t1
print("SELFTIMED:", t)
```

```python
# unpacks the resulting sparse matrix
def get_matches_df(sparse_matrix, name_vector, top=100):
    non_zeros = sparse_matrix.nonzero()
    
    sparserows = non_zeros[0]
    sparsecols = non_zeros[1]
    
    if top:
        nr_matches = top
    else:
        nr_matches = sparsecols.size
    
    left_side = np.empty([nr_matches], dtype=object)
    right_side = np.empty([nr_matches], dtype=object)
    similairity = np.zeros(nr_matches)
    
    for index in range(0, nr_matches):
        left_side[index] = name_vector[sparserows[index]]
        right_side[index] = name_vector[sparsecols[index]]
        similairity[index] = sparse_matrix.data[index]
    
    return pd.DataFrame({'left_side': left_side,
                          'right_side': right_side,
                           'similairity': similairity})
```

```python
# store the  matches into new dataframe called matched_df
matches_df = get_matches_df(matches, player_names_df, top=19276)
matches_df = matches_df[matches_df['similairity'] < 0.99999] # For removing all exact matches
matches_df = matches_df.loc[matches_df.groupby('right_side')['similairity'].agg(pd.Series.idxmax)]#For get best similarity
```

```python
left_dict = pd.Series(matches_df.left_side.values,index=matches_df.right_side).to_dict()
target_left['Player_right'] = target_left['Player'].map(left_dict)
```

```python
#Exact Name Match
df_concat = pd.merge(df_left, target_left, left_on='long_name', right_on = 'Player_right', how='left')
df_concat2 = df_concat[~df_concat['Player'].isna()]
df_left2 = df_concat[df_concat['Player'].isna()]
df_left2 = df_left2.iloc[:,:-3]
#df_concat2 = df_concat2.iloc[:,:-4]
print('Number of matched players:',len(df_concat2))
print('Number of players left in main:',len(df_left2))
df_concat2 = df_concat2.iloc[:,:-2]
df_temp = pd.merge(df_concat2, target_left, left_on='Player', right_on = 'Player', how='right',indicator=True)
target_left2 = df_temp[df_temp['_merge']=='right_only'][['Player','Rating']]
df_concat2 = df_concat2.iloc[:,:-1]
print('Number of players left in target:',len(target_left2))
```
